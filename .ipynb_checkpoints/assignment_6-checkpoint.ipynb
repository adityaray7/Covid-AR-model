```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math
from statsmodels.tsa.ar_model import AutoReg as AR
import sklearn.metrics as skm
covid_data = pd.read_csv("/Users/rajsinghbani/Downloads/LAB6-2022/daily_covid_cases.csv")
covid_data.Date = pd.to_datetime(covid_data.Date)
display(covid_data)
```

```python
ax = sns.lineplot(
    x='Date',
    y='new_cases',
    data= covid_data)
ax.set_xticklabels(['Feb-20','Apr-20','Jun-20','Aug-20','Oct-20','Dec-20','Feb-21','Apr-21','Jun-21','Aug-21','Oct-21'])
plt.show()
```

```python
covid_data_1lag = covid_data.set_index(["Date"]).shift(1)
covid_data['lagged_one'] = covid_data_1lag['new_cases'].values
display(covid_data.corr())
plt.scatter(covid_data.new_cases, covid_data.lagged_one)
plt.show()
```

```python
plt.figure()
a_corr = []
for i in range(1,7):
    plt.subplot(2,3,i)
    covid_data_ilag = covid_data.set_index(["Date"]).shift(i)
    covid_data['lagged-i'] = covid_data_ilag['new_cases'].values
    plt.scatter(covid_data.new_cases, covid_data['lagged-i'])
    plt.title(f'p = {i}')
    a_corr.append(covid_data.new_cases.corr(covid_data['lagged-i']))
plt.show()
sns.lineplot(x=[1,2,3,4,5,6],y=a_corr)
plt.xlabel('p value')
plt.ylabel('Auto-Correlation Value')
plt.show()
```

```python
#question 2
covid_data = pd.read_csv("/Users/rajsinghbani/Downloads/LAB6-2022/daily_covid_cases.csv",parse_dates=['Date'],index_col=['Date'])
X = covid_data.values
test_size = math.ceil(len(X)*0.35)
cov_train, cov_test = X[:len(X)-test_size], X[len(X)-test_size:]
def predictions(train, test, p, pred, coef):
    train_ = train[p:]
    train_ = [train_[i] for i in range(len(train_))]
    for j in range(len(test)):
        yhat = coef[0]
        for h in range(p):
            yhat += coef[h+1] * train_[len(train_)-1-h]
        pred.append(yhat)
        train_.append(test[j])
    return pred
model = AR(cov_train, lags = 5).fit()
coef_p5 = model.params
pred_p5 = []
pred_p5 = predictions(cov_train, cov_test, 5, pred_p5, coef_p5)
plt.scatter(y = pred_p5, x = cov_test)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()

plt.plot(cov_test, label = 'Actual')
plt.plot(pred_p5, label = 'Predicted')
plt.xlabel('Time')
plt.ylabel('New Cases')
plt.legend()
plt.show()

rmspe_p5 = (np.sqrt(skm.mean_squared_error(cov_test, pred_p5))*100)/cov_test.mean()
mape_p5 = skm.mean_absolute_percentage_error(cov_test, pred_p5)
print(rmspe_p5, mape_p5)
```

```python
pvalues = [1,5,10,15,25]
rmspe = []
mape = []
for p in pvalues:
    pred = []
    model_ = AR(cov_train, lags = p).fit()
    coef = model_.params
    pred = predictions(cov_train, cov_test, p, pred, coef)
    rmspe.append((np.sqrt(skm.mean_squared_error(cov_test, pred))*100)/cov_test.mean())
    mape.append(skm.mean_absolute_percentage_error(cov_test, pred))

x = np.arange(len(pvalues))
plt.yscale('log')
plt.bar(x - 0.2, rmspe, 0.4, label='rmspe')
plt.bar(x + 0.2, mape, 0.4, label='mape')
plt.xticks(x, pvalues)
plt.legend()
plt.ylabel('Error')
plt.xlabel('Lagged_Values')
plt.show()
```

```python
sat = False
p = 0
lim = 2/np.sqrt(len(cov_train))
x = covid_data.copy()
while(sat == False):
    p += 1
    j = covid_data.shift(p)
    covid_data['new'] = j.values
    corr = covid_data.new_cases.corr(covid_data['new'])
    if np.abs(corr)<=lim:
        sat = True
    covid_data = x
p -= 1
model = AR(cov_train, p).fit()
coef = model.params
pred_final = []
pred_final = predictions(cov_train, cov_test, p, pred_final, coef)
rmspe_fin = (np.sqrt(skm.mean_squared_error(cov_test, pred_final))*100)/cov_test.mean()
mape_fin = skm.mean_absolute_percentage_error(cov_test, pred_final)
print(rmspe_fin, mape_fin)
print(rmspe[3], mape[3])
```
